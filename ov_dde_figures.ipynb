```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set('talk', 'whitegrid', font_scale=1.5, rc={"lines.linewidth": 2, 'grid.linestyle': '-'})
```

```python
MaxDist20_p64_1 = np.load('ov_N20_p64_r1.0.npy')
MaxDist20_p64_2 = np.load('ov_N20_p64_r2.0.npy')
MaxDist20_p64_3 = np.load('ov_N20_p64_r3.0.npy')
MaxDist20_p64_4 = np.load('ov_N20_p64_r4.0.npy')
MaxDist40_p48_1 = np.load('ov_N40_p48_r2.0.npy')
MaxDist40_p48_2 = np.load('ov_N40_p48_r4.0.npy')
MaxDist40_p48_3 = np.load('ov_N40_p48_r6.0.npy')
MaxDist40_p48_4 = np.load('ov_N40_p48_r8.0.npy')
MaxDist80_p48_1 = np.load('ov_N80_p48_r4.0.npy')
MaxDist80_p48_2 = np.load('ov_N80_p48_r8.0.npy')
MaxDist80_p48_3 = np.load('ov_N80_p48_r12.0.npy')
```
